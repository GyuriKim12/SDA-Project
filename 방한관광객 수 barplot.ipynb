```python
import seaborn as sns
import pandas as pd
import matplotlib as mpl
mpl.rcParams['font.family']='NanumGothic'

data_1=pd.read_csv("2021_1_방한여행(국적별).csv")
data_2=pd.read_csv("2021_2_방한여행(국적별).csv")
data_3=pd.read_csv("2021_3_방한여행(국적별).csv")
data_4=pd.read_csv("2021_4_방한여행(국적별).csv")
data_5=pd.read_csv("2021_5_방한여행(국적별).csv")
data_6=pd.read_csv("2021_6_방한여행(국적별).csv")
data_7=pd.read_csv("2021_7_방한여행(국적별).csv")
data_8=pd.read_csv("2021_8_방한여행(국적별).csv")
data_9=pd.read_csv("2021_9_방한여행(국적별).csv")
data_10=pd.read_csv("2021_10_방한여행(국적별).csv")
data_11=pd.read_csv("2021_11_방한여행(국적별).csv")
data_12=pd.read_csv("2021_12_방한여행(국적별).csv")
```

```python
data_1['월']="1월"
data_2['월']="2월"
data_3['월']="3월"
data_4['월']="4월"
data_5['월']="5월"
data_6['월']="6월"
data_7['월']="7월"
data_8['월']="8월"
data_9['월']="9월"
data_10['월']="10월"
data_11['월']="11월"
data_12['월']="12월"
```

```python
data=pd.concat([data_1,data_2,data_3,data_4,data_5,data_6,data_7,data_8,data_9,data_10,data_11,data_12,],axis=0)
data.reset_index(drop=True,inplace=True)
```

```python
#월별 방한관광객 수 추이 bar그래프

sns.barplot(x="월", y="방한관광객", errorbar=('ci', False), data=data)
```

```python
usa = data[data["국적"]=="미국"]["방한관광객"].mean()
pillipin = data[data["국적"]=="필리핀"]["방한관광객"].mean()
china = data[data["국적"]=="중국"]["방한관광객"].mean()
miyanma = data[data["국적"]=="미얀마"]["방한관광객"].mean()
indonasia = data[data["국적"]=="인도네시아"]["방한관광객"].mean()
canada = data[data["국적"]=="캐나다"]["방한관광객"].mean()
```

```python
#국가 별 방한관광객 수 bar 그래프

mean = [usa, pillipin, china, miyanma, indonasia, canada]
country = ["미국", "필리핀", "중국", "미얀마", "인도네시아", "캐나다"]

df = pd.DataFrame({"국가":country, "방한관광객 수":mean})
df = df.sort_values(by=["방한관광객 수"], ascending=False)
df.plot.bar(x="국가", y="방한관광객 수")
```
