```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt 
```

```python
data_1=pd.read_csv("2021_1_미국 방한목적.csv")
data_2=pd.read_csv("2021_2_미국 방한목적.csv")
data_3=pd.read_csv("2021_3_미국 방한목적.csv")
data_4=pd.read_csv("2021_4_미국 방한목적.csv")
data_5=pd.read_csv("2021_5_미국 방한목적.csv")
data_6=pd.read_csv("2021_6_미국 방한목적.csv")
data_7=pd.read_csv("2021_7_미국 방한목적.csv")
data_8=pd.read_csv("2021_8_미국 방한목적.csv")
data_9=pd.read_csv("2021_9_미국 방한목적.csv")
data_10=pd.read_csv("2021_10_미국 방한목적.csv")
data_11=pd.read_csv("2021_11_미국 방한목적.csv")
data_12=pd.read_csv("2021_12_미국 방한목적.csv")
```

```python
data_1['월']="1월"
data_2['월']="2월"
data_3['월']="3월"
data_4['월']="4월"
data_5['월']="5월"
data_6['월']="6월"
data_7['월']="7월"
data_8['월']="8월"
data_9['월']="9월"
data_10['월']="10월"
data_11['월']="11월"
data_12['월']="12월"
```

```python
data=pd.concat([data_1,data_2,data_3,data_4,data_5,data_6,data_7,data_8,data_9,data_10,data_11,data_12,],axis=0)
data.reset_index(drop=True,inplace=True)
```

```python
trip = data[data['구분']==data['구분'][0]]
study = data[data['구분']==data['구분'][1]]
public = data[data['구분']==data['구분'][2]]
comm = data[data['구분']==data['구분'][3]]
```

```python
#월별 데이터 집단으로 모아서 관광, 어학연수, 공용, 상용 간 차이 분석
from scipy.stats import shapiro
print(shapiro(trip['비율']))
print(shapiro(study['비율']))
print(shapiro(public['비율']))
print(shapiro(comm['비율']))
#p-value 값이 0.05보다 큰 경우가 존재해 정규성을 만족하지 않는다
#-> anova 분석, 사후분석 불가능
#bar그래프로 시각화해서 분포 차이만 눈으로 보여주기..?
```

```python
#관광 목적의 미국인 관광객 비율
trip_data = data[data['구분']==data['구분'][0]]
trip_data = trip_data.reset_index(drop=True)
trip_data
plt.bar(trip_data['월'], trip_data['비율'])
plt.show()
```

```python
# 미국인 관광객의 방한 목적 비율
x = ["관광", "어학연수", "공용", "상용"]
y = [trip['비율'].mean(), study['비율'].mean(), public['비율'].mean(), comm['비율'].mean()]
plt.bar(x, y)
plt.show()
```
