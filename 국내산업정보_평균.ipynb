```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
from scipy import stats
```

```python
data = pd.read_csv("국내산업정보.csv",encoding = "CP949")
data
```

```python
df=data[['산업명','2016년','2017년','2018년','2019년','2020년']]
df
```

```python
mean_data=df.mean()
mean_data
```

```python
import matplotlib as mpl
mpl.rcParams['font.family'] = 'NanumGothic'
ax=mean_data.plot(kind='bar', title='국내산업정보')
ax.set_xlabel("년도")
ax.set_ylabel("평균")
```
